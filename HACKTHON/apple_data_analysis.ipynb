```python
def remove_outliers(dataframe):
    result = dataframe.copy()
    columns_list = list(result.columns)
    for column in columns_list:
        if result[column].dtype in [int, float]:
            q1 = result[column].quantile(0.25)
            q3 = result[column].quantile(0.75)

            iqr = q3 - q1
            lower_fence = q1 - iqr*1.5
            upper_fence = q3 + iqr*1.5
            result = result[(result[column] >= lower_fence) &
                                    (result[column] <= upper_fence)]

    # Reset index after removing outliers
    result.reset_index(drop=True, inplace=True)
    return result
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(input_data, output_data, test_size=0.2, random_state=100)
```

```python
# Instantiate the KNN Classifier
knn = KNeighborsClassifier(n_neighbors=5)  # You can adjust the number of neighbors (k)

# Model Training
knn.fit(X_train, y_train)

# Model Evaluation
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
classifier = KNeighborsClassifier()
classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = classifier.predict(X_test)

# Generate and print the classification report
report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)
```

```python
report = classification_report(y_test, y_pred, output_dict=True)

# Convert the classification report to a DataFrame
df_report = pd.DataFrame(report).transpose()

# Plotting the filled chart
plt.figure(figsize=(8, 6))
sns.heatmap(df_report.iloc[:-1, :-1], annot=True, cmap="Blues", fmt=".2f", linewidths=.5)
plt.title('Classification Report - KNN')
plt.show()
```

```python
# Instantiate Random Forest classifier
random_forest = RandomForestClassifier(n_estimators=100, random_state=100)  # You can adjust the number of trees (n_estimators)

# Train the Random Forest classifier
random_forest.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = random_forest.predict(X_test)

# Evaluate the performance of Random Forest
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from scipy.stats import norm

def calculate_p_value(confidence_interval):
    # Assuming a two-tailed test
    alpha = 1 - confidence_interval
    z_score = norm.ppf(1 - alpha / 2)

    # Calculate p-value
    p_value = 2 * (1 - norm.cdf(z_score))

    return p_value

# Example: 95% confidence interval
confidence_interval = 0.95
p_value_95 = calculate_p_value(confidence_interval)

print(f'P-value for a {confidence_interval * 100}% confidence interval: {p_value_95}')
```
